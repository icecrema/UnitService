## Read the data, pre calculate the daily census

```python
# read data from file and calculate the census, adjust the census
# census_file = "../Data/2019-09-13Census and Surgical Admits and Scheduled 2.csv"
census_file = "../Data/2020-01-10Census and Surgical Admits and Scheduled 2.csv"

# fyear = 2020
# nogrowth = True
# start_date = datetime.datetime(year = fyear-1, month = 9, day = 1)
# end_date = datetime.datetime(year = 2020, month = 1, day = 1)

fyear = 2019
nogrowth = False
start_date = datetime.datetime(year = fyear-1, month = 9, day = 1)
end_date = datetime.datetime(year = 2019, month = 9, day = 1)

units_to_consider = ['PCU160', 'PCU200', 'PCU300', 'PCU360', 'PCU380', 'PCU400', 'PCU500', 'PCU520']
services_to_consider = ['HemeOnc and StemCell', 'General Surgery', 'Cardiology', 'Green Team',
                         'Red Team',
                         'Urology', 'Orthopedics', 'Pulmonary',
                          'Neurosurgery',
                          'Neurology',
                          'Otolaryngology (ENT)',
                          'Plastic Surgery', 'HemeOnc and StemCell']

data = CensusData(fyear=fyear, units_to_consider = units_to_consider, 
                              services_to_consider = services_to_consider)
data.read_census_from_file(census_file)
data.data_preprocess(start_date = start_date, end_date = end_date)
data.filter_units_services()
data.cal_daily_census()
data.daily_census_adjust(nogrowth = nogrowth)
data.divide_gen_peds() # divide the gen peds into two teams equally
```

## Build the optimization Model

```python
units_to_consider = data.get_units_to_consider()
services_to_consider = data.get_services_to_consider()
service_count_daily = data.get_adjusted_daily_census()
unit_cap_dict = data.get_dict_unit_cap()
print(units_to_consider)
print(services_to_consider)
```

```python
unit_cap_dict
```

```python
# run the model
allocation = constrained_qp(services_to_consider, units_to_consider, service_count_daily,
                    unit_cap_dict, cap_thresh = 0.90)
```

```python
# run the model
allocation = constrained_piecewise(services_to_consider, units_to_consider, service_count_daily,
                    unit_cap_dict, cap_thresh = 0.90)
```

```python
# print the allocation
temp = allocation_to_dict(allocation, units_to_consider, services_to_consider)
for unit in temp:
    print(unit, temp[unit])
```

## Evaluate the optimal allocation

```python
allocation_dict = allocation_to_dict(allocation, units_to_consider, services_to_consider)
present_performance(allocation_dict, service_count_daily, 
                        unit_cap_dict,
                        units_to_consider = ['PCU200', 'PCU300', 'PCU360', 'PCU400', 'PCU500'], 
                        cap_thresh = 0.9, 
                        over_cap_transfer = 'floor')
```
